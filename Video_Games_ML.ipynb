### ***LINEAR REGRESSION***

### ***EXPLANATION : What is Linear Regression?***

> : Linear Regression is a supervised machine learning algorithm used to predict a continuous numerical value (such as price, height, temperature, or marks) based on one or more input features.

> :It assumes a linear relationship between the independent variable(s) and the dependent variable.

*Mathematical Model*
**For a single feature:**
*ùë¶=ùëöùë•+ùëè*

****For multiple features:****
ùë¶=ùëè0+ùëè1ùë•1+ùëè2ùë•2+‚ãØ+ùëèùëõùë•ùëõ

>  : Where:

> : y = predicted output

> : x = input feature(s)

> : ùëè0 = intercept

> : ùëèùëõ= coefficients (weights)

How Linear Regression Works :

***1.Takes input features ùëã***

***2.Predicts output using a linear equation***

***3.Calculates the error (difference between predicted and actual values)***

***4.Adjusts coefficients to minimize error (usually using Mean Squared Error)***

> For Ex : > from sklearn.linear_model import LinearRegression

> model = LinearRegression()

> model.fit(X_train, y_train)

> predictions = model.predict(X_test)


```python
#appling the data into X and Y.
X = data['Release Year']
y = data['Price']
```

```python
# importing train_test_split model to regonize the arruracy , prediction 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
# Let's now take a look at the X_train dataset
X_train.head().reset_index()
```

```python
# Let's now take a look at the y_train dataset 
y_train.head().reset_index()
```

#### Building a Linear Model

You first need to import the `statsmodel.api` library using which you'll perform the linear regression.

```python
import statsmodels.api as sm
```

By default, the `statsmodels` library fits a line on the dataset which passes through the origin. But in order to have an intercept, you need to manually use the `add_constant` attribute of `statsmodels`.

```python
# Add a constant to get an intercept
X_train_sm = sm.add_constant(X_train)

# Fit the resgression line using 'OLS'
lr = sm.OLS(y_train, X_train_sm).fit()
```

```python
# Print the parameters, i.e. the intercept and the slope of the regression line fitted
lr.params
```

```python
# Performing a summary operation lists out all the different parameters of the regression line fitted
print(lr.summary())
```

```python
y_train_pred = lr.predict(X_train_sm)
res = (y_train - y_train_pred)
```

###  Prediction on the Test Set*

Now that you have fitted a regression line on your train dataset, it's time to make some predictions on the test data.

```python
# Add a constant to the data
X_test_sm = sm.add_constant(X_test)

#Predcit the y value corresponding X_test_sm
y_pred = lr.predict(X_test_sm)
```

```python
#checking y_pred 
y_pred.head()
```

```python
#importing mean_square_error and r2_score from sklearn
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
# Looking at the rmse (return to mean squared error)
# we will take sqrt (square root)
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
# checking the R-squared on the test set
r_squared  = r2_score(y_test,y_pred)
r_squared
```

###  ***LOGISTIC AGRESSION***

***What is Logistic Regression?***

> Logistic Regression is a supervised classification algorithm used to predict binary outcomes (0 or 1, Yes or No, True or False).
Despite its name, it is used for classification, not regression.

***Mathematical Model***

### Logistic Regression uses the sigmoid function

> œÉ(z)=1
    1+e‚àíz1

> Where:  z=b0‚Äã+b1‚Äãx1‚Äã+b2‚Äãx2‚Äã+‚ãØ+bn‚Äãxn‚Äã

> The output is a probability between 0 and 1.

~~ **How Logistic Regression Works**

***1.Computes a linear combination of inputs***

***2.Applies the sigmoid function***

***3.Converts output into probability***

***4.Uses a threshold (usually 0.5) to classify:***

> : ‚â• 0.5 ‚Üí Class 1

> : < 0.5 ‚Üí Class 0


***FOR EXAMPLE : from sklearn.linear_model import LogisticRegression***
> model = LogisticRegression()

> model.fit(X_train, y_train)

> predictions = model.predict(X_test)



```python
# inputing the x and y data for logistic regression . 
X = data[['User Rating','Price','Game Length (Hours)',]]
y = data['Multiplayer']
```

```python
# whichever columns contain Yes Or NO data will be chnages as 1,0. 
y = y.map({'Yes':1,'No':0})
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 0)
```

```python
# importing logistic regression model
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train,y_train)
```

### ***Accuracy Score ‚Äì Explanation***

Accuracy Score measures how many predictions made by a model are correct out of the total predictions.

In simple words, it tells us how accurate the model is.

***In Logistic Regression***

Logistic Regression is a classification algorithm.
Accuracy score checks how well the model correctly classifies data into classes (like Yes/No, 0/1). 

Example

If a model makes 100 predictions and 85 are correct:

Accuracy=85
       ------
        100
         = 0.85(85%)

```python
# after predict will check the accuracy score
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
print("Accuracy:",accuracy_score(y_test,y_pred))
```

### ***Confusion Matrix ‚Äì Explanation***

A Confusion Matrix is a table used to evaluate the performance of a classification model by comparing actual values with predicted values.

It shows where the model is correct and where it makes mistakes.

***> Meaning of Each Term***

1.True Positive (TP): Correctly predicted positive cases

2.True Negative (TN): Correctly predicted negative cases

3.False Positive (FP): Predicted positive but actually negative

4,False Negative (FN): Predicted negative but actually positive

```python
# importing confusion matrix 
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
# after all predict , accruarcy score in the last we make a classification report 
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

### ***ROC Curve (Receiver Operating Characteristic Curve)***

The ROC Curve is a graphical plot that shows the performance of a classification model at different threshold values.

> It plots:

1.X-axis: False Positive Rate (FPR)

2.Y-axis: True Positive Rate (TPR / Recall)

3.The curve shows how well the model can separate positive and negative classes.

> Key Idea:
The closer the curve is to the top-left corner, the better the model.



### ***ROC-AUC Score (Area Under the Curve)***

ROC-AUC Score is a single number that represents the area under the ROC curve.

It measures how well the model can distinguish between classes.

> Score Range

: 1.0 ‚Üí Perfect classifier

: 0.5 ‚Üí Random guessing

: < 0.5 ‚Üí Worse than random

```python
# importing roc curve and roc auc score
from sklearn.metrics import roc_curve , roc_auc_score
```

```python
# predicting
y_prob = model.predict_proba(X_test)[:,1]
```

```python
fpr,tpr,thresholds = roc_curve(y_test,y_prob) 
```

```python
auc_score = roc_auc_score(y_test,y_prob)
print("AUC Score:" , auc_score)
```

```python
# plot graph for roc curve
plt.figure(figsize=(6,6))
plt.plot(fpr , tpr , label = f"ROC Curve (AUC={auc_score:.2f})")
plt.plot([0,1],[0,1],linestyle = '--',label = 'Random Guess')

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC CURVE - MULTIPLAYER PREDICTION")
plt.legend()
plt.show()
```
